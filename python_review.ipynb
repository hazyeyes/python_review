Exercise 1: Temperature Converter
Create a program that converts temperatures between Celsius and Fahrenheit.
Instructions:
1.	Ask the user to input a temperature.
2.	Ask the user to select the conversion type: from Celsius to Fahrenheit or from Fahrenheit to Celsius.
3.	Perform the appropriate conversion and print the result.

```python
def ask_temperature():
    try:
        temperature = float(input("Enter the temperature: "))
    except ValueError:
        print("Please enter an integer...")
        print("--------------------------")
        print()
        main()
    return temperature

def additional():
    ask_user = input("Do you want to convert again? (y/n): ")
    if ask_user.lower() == "y":
        print("______________________")
        print()
        main()
    elif ask_user.lower() == "n":
        quit()
    else:
        print("Enter \"y\" or \"n\".")
        additional()


def main():
        print("Temperature Converter")
        print("")
        temperature = ask_temperature()
        conversion_type = input("Select the conversion type \nA. Celsius to Fahrenheit \nB. Fahrenheit to Celsius \n(\"A\" or \"B\"): ")

        if conversion_type.upper() == "A":
            temperature = "{:.2f}".format((temperature * (9/5)) + 32)
            print("----------------------")
            print("Temperature: ", temperature, u"\u2109")
            additional()
        elif conversion_type.upper() == "B":
            temperature = "{:.2f}".format((temperature - 32) * (5/9))
            print("----------------------")
            print("Temperature: ", temperature, u"\u2103")
            additional()
        else:
            print("Please enter a valid input: ")
            print("----------------------------")
            print()
            main()

main()

    
```


Exercise 2: Ohmâ€™s Law Calculator
Instructions:
1.	Ask the user what they want to calculate: Voltage, Current, or Resistance.
2.	Based on their choice, prompt the user to input the appropriate values.
3.	Use Ohm's Law to calculate the missing variable and display the result.
4.	Handle cases where division by zero might occur.

```python
def user_inp():
    comp = {"V","I","R"}
    user_input = input("What do you want to solve?\n(\"V\" for Voltage | \"I\" for Current | \"R\" for Resistance)\n: ")
    user_input = user_input.upper()

    if user_input not in comp:
        print("Please enter a valid value.")
        print("--------------------------")
        print("")
        main()
    
    return user_input

def multiply(x,y):
    answer = x * y
    return answer

def divide(x,y):
    if y != 0:         
        answer = x / y
        return answer
    else:
        print(f"Please enter a valid given")
        print("--------------------------")
        print("")
        main()

def additional():
    ask_user = input("Do you want to convert again? (y/n): ")
    if ask_user.lower() == "y":
        print("______________________")
        print()
        main()
    elif ask_user.lower() == "n":
        quit()
    else:
        print("Enter \"y\" or \"n\".")
        additional()

def main():
    print("Ohm's Law Calculator")
    print("     / \     ")
    print("    / V \    ")
    print("   /-----\   ")
    print("  / I | R \  ")
    print(" ----------- ")

    user_input = user_inp()

    if user_input == "V":
        print("Enter the necessary values")
        current = float(input("Current (I): "))
        resistance = float(input("Resistance (R): "))
        result = "{:.2f}".format(multiply(current, resistance))
        print("--------------------------")
        print("The Voltage is ", result, f"Volts \n(Current: {current} Amperes, Resistance: {resistance} Ohms)")
        additional()  
    elif user_input == "I":
        print("Enter the necessary values")
        voltage = float(input("Voltage (V): "))
        resistance = float(input("Resistance (R): "))
        result = "{:.2f}".format(divide(voltage, resistance))
        print("--------------------------")
        print("The Current is ", result, f"Amperes \n(Voltage: {voltage} Volts, Resistance: {resistance} Ohms)")  
        additional()
    elif user_input == "R":
        print("Enter the necessary values")
        voltage = float(input("Voltage (V): "))
        current = float(input("Current (I): "))
        result = "{:.2f}".format(divide(voltage, current))
        print("--------------------------")
        print("The Resistance is ", result, f"Ohms \n(Voltage {voltage} Volts, Current: {current} Amperes)")  
        additional()
    else:
        print("Please enter a valid value.")
        main()

main()
```

Exercise 3:  Diamond Shape (advance topic):

Write a Python function named print_diamond that takes an odd integer n as an argument and prints a diamond shape with a width of n using the * character.

```python
def int_input():
    try:
        inp = int(input("Please enter an odd number: "))
    except ValueError:
        print("Please enter an integer")
        print("--------------------------")
        print("")
        main()
    return inp

def additional():
    ask_user = input("Do you want to convert again? (y/n): ")
    if ask_user.lower() == "y":
        print("______________________")
        print()
        main()
    elif ask_user.lower() == "n":
        quit()
    else:
        print("Enter \"y\" or \"n\".")
        additional()

def main():
    print("Print a Diamond.")    
    print("----------------")

    user_input = int_input()
    if (user_input % 2) != 0:
        for i in range(user_input):
            print(" " * (user_input - i - 1) + "*" * (2 * i + 1))
        for i in range(user_input - 2, -1, -1):
            print(" " * (user_input - i - 1) + "*" * (2 * i + 1))
        additional()
    else:
        print("Please enter an odd integer.")
        print("--------------------------")
        print("")
        main()

main()
```
